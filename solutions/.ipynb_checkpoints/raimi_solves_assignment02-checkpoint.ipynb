### Preprocessing

1. Convert rgb to grayscale
2. Detect edges with canny edge detector

```python
LANE_INPUT_ROOT = "images/lane_detection/"
LANE_CANNY_ROOT = "solutions/lane_detection/canny"

if not os.path.exists(LANE_CANNY_ROOT):
    os.makedirs(LANE_CANNY_ROOT)

for input_file in sorted(glob.glob(os.path.join(LANE_INPUT_ROOT,"*"))):
    filename = os.path.basename(input_file)
    lane_rgb = io.imread(input_file)
    lane_gray = color.rgb2gray(lane_rgb)
    edges = feature.canny(lane_gray,sigma=2)
    io.imsave(os.path.join(LANE_CANNY_ROOT,filename), img_as_ubyte(edges) )
```

### Hough Transform

```python
MAX_ALPHA = math.pi

def hough_transform(edges, n_alpha, n_r):
    MAX_Y, MAX_X = edges.shape
    MAX_R = math.hypot(MAX_X, MAX_Y)
    
    ALPHA_VALUES = np.linspace(0, MAX_ALPHA, n_alpha)
    R_VALUES = np.linspace(-MAX_R, MAX_R, n_r)
    
    accumulator = np.zeros((n_r, n_alpha))
    for y, x in zip(*edges.nonzero()):
        for alpha_idx, alpha in enumerate(ALPHA_VALUES):
            r = x * math.cos(alpha) + y * math.sin(alpha)
            r_idx = int(( (r+MAX_R) / (2*MAX_R) ) * n_r)
            accumulator[r_idx, alpha_idx] += 1
    return accumulator, R_VALUES, ALPHA_VALUES
```

### Convert Image to Parameter Space

```python
lane_edges = io.imread(os.path.join(LANE_CANNY_ROOT,"f00000.png"))

N_ALPHA = 1000
N_R = 1000

acc, r_values, alpha_values = hough_transform(lane_edges, N_ALPHA, N_R)
```

### Find 10 Strongest Lines
This is equivalent to finding peaks in parameter space.

```python
N_LINES = 10
strongest_lines = feature.peak_local_max(acc, min_distance=15, num_peaks=N_LINES)
```

### Visualization of the Parameter Space

```python
plt.figure(figsize=(6,6))
plt.xlabel("alpha (radians)")
plt.ylabel("r (pixel)")
plt.xticks(np.arange(alpha_values.size)[::200],np.around(alpha_values[::200],2) )
plt.yticks(np.arange(r_values.size)[::100],np.around(r_values[::100],0) )
plt.imshow(acc, cmap="gray")
plt.plot(strongest_lines[:, 1], strongest_lines[:, 0], 'rx',ms=6, mew=1.5)
plt.show()
```

### Visualization of 10 Strongest Lines

```python
plt.figure(figsize=(14,6))
plt.imshow(lane_edges, cmap=plt.cm.gray)
for r_idx, alpha_idx in strongest_lines:
    r = r_values[r_idx]
    alpha = alpha_values[alpha_idx]
    (x0, y0) = r * np.array([np.cos(alpha), np.sin(alpha)])
    plt.axline((x0, y0), slope=np.tan(alpha + np.pi/2),c="r",ls="--")
    plt.axis(( 0,lane_edges.shape[1],lane_edges.shape[0],0))
```
