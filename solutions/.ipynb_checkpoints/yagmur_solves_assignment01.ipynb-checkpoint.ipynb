```python
# imports
%matplotlib inline
from skimage import io, data, color
import matplotlib.pyplot as plt
import matplotlib.image as mpimg
import numpy as np
```

```python
def conv2d(image, kernel):
    # here goes your code
    m, n = kernel.shape
    #überprüfen ob nxn-matrix
    if (m == n):
        #bestimmen vektor vom bild für Länge beider Achsen
        y, x = image.shape
        y = y - m + 1
        x = x - m + 1
        # bilden neues Array mit neuer Länge der Achsen und nur aus nullen
        new_image = np.zeros((y,x))
        # gehen jeden pixel im neuen Array durch
        for i in range(y):
            for j in range(x):
                # setzen für Nullen neue Werte ein, die aus der Summe der Multiplikation des Bildes und Kernels berechnet werden
                new_image[i][j] = np.sum(image[i:i+m, j:j+m]*kernel)
    return new_image

    #result = image # remove this please
    #return result

image = color.rgb2gray(io.imread('dog.jpg'))
kernel = np.array([[1, 1, 1], [1, 1, 1], [1, 1, 1]])
plt.imshow(conv2d(image, kernel), cmap='gray')

#kernel2 = np.array([[-1, 0, 1], [-1, 0, 1], [-1, 0, 1]])
#plt.imshow(conv2d(image, kernel2), cmap='gray')
```

```python
kernel2 = np.array([[-1, -2, -1], [0, 0, 0], [1, 2, 1]])
output1 = conv2d(image, kernel2)
plt.imshow(conv2d(image, kernel2), cmap='gray')
```

```python
kernel3 = np.array([[-1, 0, 1], [-2, 0, 2], [-1, 0, 1]])
output2 = conv2d(image, kernel3)
plt.imshow(conv2d(image, kernel3), cmap='gray')
```
